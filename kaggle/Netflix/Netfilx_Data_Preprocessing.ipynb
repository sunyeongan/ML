```python

cnt = 0
NanCnt = 0
main_df['rating'].isna().values.any()
print(main_df.shape)
for i in main_df.iloc[:,8]: #rating 열만 출력
    if pd.isna(i) == True: #결측값이 존재한다면
        NanCnt+=1
        temp = main_df['rating'].sample(n=1,replace=True)
        main_df['rating'].fillna(temp) # !!!!
    print(cnt,':',i)
    cnt=cnt+1
print(NanCnt)
```

```python
#nan 처리방법 
#1. 데이터에서 삭제 .dropna()
#2. 다른값으로 치환 fillna()

# Q. 2번을 사용하여,딕셔너리를 이용하지않고, 결측값에 rating값을 랜덤하게 넣는다 
#NaN 값을 랜덤값으로 대체하기
temp = main_df['rating'].sample(n=1,replace=True)

print(temp)#복원으로 1개 추출

#main_df['rating'].fillna('hhh') #복원으로 1개 추출)

```

```python
rating_replacements = { 67: 'TV-PG', 
                       2359: 'TV-14', 
                       3660: 'TV-MA', 
                       3736: 'TV-MA', 
                       3737: 'NR', 
                       3738: 'TV-MA', 
                       4323: 'TV-MA ' }


for id, rate in rating_replacements.items():
    main_df.iloc[id, 8] = rate
    print(main_df.iloc[id,8])#비어있던 7개의 컬럼에 값이 채워짐
    
main_df['rating'].isna().sum()
```

```python
# 결측치가 많은 director,cast 컬럼은 삭제한다.
```

```python


#main_df = main_df.drop(['director', 'cast'], axis=1)
#main_df.columns
main_df = main_df.drop([main_df.index[3],main_df.index[4]]) 
main_df.columns
```

```python
#컬럼값을 채우기 힘든 date_added 컬럼은 빈데이터가 있는 행을 삭제한다.
```

```python
main_df = main_df[main_df['date_added'].notna()]
```

```python
main_df.isna().sum()
```

```python
#.fillna() -> 결측값을 최빈값(가장 많이 관측되는 수)으로 채우기 
#.mode() -> 최빈값 
main_df['country'] = main_df['country'].fillna(main_df['country'].mode()[0])

```

```python
#데이터 전치리 완료
main_df.isna().sum()
```

```python
# date_added값에서 year,month 을 추출해서 컬럼에 추가한다.

#year_added 컬럼 추가
main_df['year_added'] = main_df['date_added'].apply(lambda x: x.split(" ")[-1])

#month_added 컬럼 추가

main_df['month_added'] = main_df['date_added'].apply(lambda x: x.split(" ")[0])


#" "으로 분리

print(main_df['date_added'])

#print(main_df['date_added'].apply(lambda x: x.split(" ")[-1]))
```

```python
#컬럼추가 확인
main_df.head()
```

```python
ratings_ages = { 'TV-PG': 'Older Kids', 
                'TV-MA': 'Adults', 
                'TV-Y7-FV': 'Older Kids',
                'TV-Y7': 'Older Kids',
                'TV-14': 'Teens',
                'R': 'Adults',
                'TV-Y': 'Kids',
                'NR': 'Adults',
                'PG-13': 'Teens',
                'TV-G': 'Kids',
                'PG': 'Older Kids',
                'G': 'Kids',
                'UR': 'Adults',
                'NC-17': 'Adults' 
}

#target_ages 컬럼 추가
#rating 컬럼을 이용해서 시청연령에 대한 값을 풀어서 저장
main_df['target_ages'] = main_df['rating'].replace(ratings_ages)
main_df['target_ages'].unique()
```

## 데이터 타입 변경

```python
#type 컬럼 category로 변경
main_df['type'] = pd.Categorical(main_df['type'])

#target_ages 컬럼 category로 변경
main_df['target_ages'] = pd.Categorical(main_df['target_ages'],categories=
                                       ['Kids','Older Kids','Tenns','Adults'])

#year_added는 integer로 변경 -> released_year 데이터와 비교 가능 
main_df['year_added'] = pd.to_numeric(main_df['year_added'])

main_df.dtypes
```
